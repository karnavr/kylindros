```julia
using Plots
using Random
```

```julia
# generate x values, e.g. uniform in log10 scale between 0 and -16
n = 100
x = range(0, -16, length=n)
y = 10.0 .^ x  # This spans 1.0 to 1e-16

# add some noise (so points are not exactly on the curve)
noise = 10.0 .^ x .* (0.1 .* (2 .* rand(n) .- 1)) # up to ±10% noise
y_noisy = y .+ noise
```

```julia
scatter(y_noisy, yscale = :log10)

# find max and min of y_noisy
max_y = maximum(y_noisy)
min_y = minimum(y_noisy)

# find the closest power of 10 to max and min
max_y_exp = floor(log10(max_y))
min_y_exp = floor(log10(min_y))

println(max_y_exp, min_y_exp)

# set yticks to be all powers of 10 between max and min
arrays = [10^i for i in max_y_exp:-2:min_y_exp]

yticks!(arrays)
```
